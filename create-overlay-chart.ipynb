```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df1 = pd.read_csv ("202106 Argon Neo Temperature Comparison - raw01.csv")
df2 = pd.read_csv ("202106 Argon Neo Temperature Comparison - raw02.csv")
```

```python
df2.columns
```

```python
fig = go.Figure()
```

```python
fig.add_trace(go.Scatter(x=df1['time:'], y=df1['neoclosenofan'],
                    mode='lines',
                    name='neoclosenofan'))
                    
fig.add_trace(go.Scatter(x=df2['time:'], y=df2['neoopennofan'],
                    mode='lines',
                    name='neoopennofan'))

fig.show()
```
